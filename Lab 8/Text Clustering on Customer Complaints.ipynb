```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer
from tabulate import tabulate
from collections import Counter
```

```python
df = pd.read_csv('customer_complaints_1.csv')
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(df['text'])
```

```python
k = 3 
km = KMeans(n_clusters=k, random_state=42)
km.fit(X)

y_pred = km.predict(X) 

# Display a sample of the document and its predicted cluster
table_data = [["Document Sample", "Predicted Cluster"]]
# Showing first 10 rows for brevity
table_data.extend([[doc, cluster] for doc, cluster in zip(df['text'], y_pred])) 
print(tabulate(table_data, headers="firstrow"))
```
